## 4. Folium map

### for 2015

```python
# Initialize a folium map
world_map = folium.Map(location=[20, 0], zoom_start=2)

# Add the choropleth layer
folium.Choropleth(
    geo_data=geojson_data,
    name="choropleth",
    data=df_2015,
    columns=["country", "happiness_score"],  # Columns from the dataset
    key_on="feature.properties.name",  # Key in GeoJSON to match with the dataset
    fill_color="YlGnBu",  # Color scale
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Happiness Score",
).add_to(world_map)

# Adding layer control
folium.LayerControl().add_to(world_map)

# To display in a Jupyter Notebook 
world_map
```

```python
# filtering pakistan data frame 
df_pak = df[df['country'] == 'Pakistan']
```

```python
# checking the columns
df_pak.columns
```

```python
# creating line chart of happiness score over the years( Pakistan)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_pak['year'], df_pak['happiness_score'], marker='o', label='Happiness Score')
plt.title('Happiness Score Over the Years (Pakistan)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
