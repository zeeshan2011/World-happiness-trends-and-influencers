## 3. Combining datasets

#### Adding a new Column 'Year' 

```python
# Add a 'year' column to each DataFrame
df_2015['year'] = 2015
df_2016['year'] = 2016
df_2017['year'] = 2017
df_2018['year'] = 2018
df_2019['year'] = 2019
```

```python
# checking columns name of each data set
print(df_2015.columns)
print(df_2016.columns)
print(df_2017.columns)
print(df_2018.columns)
print(df_2019.columns)
```

#### Standardizng column names

```python
# Standardize column names for all datasets
rename_mapping = {
    'Country or region': 'Country',
    'Happiness.Rank': 'Happiness Rank',
    'Overall rank': 'Happiness Rank',  
    'Happiness.Score': 'Happiness Score',
    'Whisker.high': 'Upper Confidence Interval',
    'Whisker.low': 'Lower Confidence Interval',
    'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
    'Health..Life.Expectancy.': 'Health (Life Expectancy)',
    'Trust..Government.Corruption.': 'Trust (Government Corruption)',
    'Score': 'Happiness Score',
    'GDP per capita': 'Economy (GDP per Capita)',
    'Social support': 'Family',
    'Healthy life expectancy': 'Health (Life Expectancy)',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Trust (Government Corruption)'
}


# Apply renaming to each dataset
df_2015.rename(columns=rename_mapping, inplace=True)
df_2016.rename(columns=rename_mapping, inplace=True)
df_2017.rename(columns=rename_mapping, inplace=True)
df_2018.rename(columns=rename_mapping, inplace=True)
df_2019.rename(columns=rename_mapping, inplace=True)
```

#### Retaining only common columns

```python
# Retain only columns that are present in all datasets
common_columns = [
    'Country', 'Happiness Rank', 'Happiness Score',
    'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
    'Freedom', 'Trust (Government Corruption)', 'Generosity', 'year'
]

df_2015 = df_2015[common_columns]
df_2016 = df_2016[common_columns]
df_2017 = df_2017[common_columns]
df_2018 = df_2018[common_columns]
df_2019 = df_2019[common_columns]
```

#### Concatenating Datasets

```python
# Concatenate the datasets
combined_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)

# Check the resulting DataFrame
print(combined_df.info())
print(combined_df.head())
```

### 4. Consistency Check

#### Checking for missing values

```python
# Check for missing values
print("\nMissing Values:")
combined_df.isnull().sum()
```

#### Checking for Duplicates

```python
# Check for duplicate rows
duplicates = combined_df.duplicated()
print(f"\nNumber of duplicate rows: {duplicates.sum()}")
```

#### Renaming Columns 

```python
# Rename columns for consistency (converting to snake_case, removing spaces, etc.)
combined_df.columns = [
    col.strip().replace(" ", "_").lower() for col in combined_df.columns
    ]
print("\nRenamed Columns:")
combined_df.columns
```
