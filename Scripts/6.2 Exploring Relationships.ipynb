## 02 Defining variables for each category

```python
# Define the variables for each category
variables_general_trends = ['year', 'happiness_score', 'happiness_rank', 'country']
variables_economic_factors = ['economy_(gdp_per_capita)', 'happiness_score', 'trust_(government_corruption)']
variables_social_health = ['family', 'happiness_score', 'health_(life_expectancy)']

# Filter the dataset for each category
df_general_trends = df[variables_general_trends]
df_economic_factors = df[variables_economic_factors]
df_social_health = df[variables_social_health]
```

```python
# Display the filtered DataFrames for verification
print("General Trends and Patterns Variables:")
print(df_general_trends.head())

print("\nEconomic Factors Variables:")
print(df_economic_factors.head())

print("\nSocial and Health Factors Variables:")
print(df_social_health.head())
```

## 03 Correlation Analysis

```python
# Combining all relevant variables for correlation analysis
correlation_columns = [
    'happiness_score', 'happiness_rank', 'economy_(gdp_per_capita)',
    'trust_(government_corruption)', 'family', 'health_(life_expectancy)'
]
correlation_data = df[correlation_columns]

# Calculating the correlation matrix
correlation_matrix = correlation_data.corr()

# Creating a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#### A strong positive correlation between happiness_score and economy_(gdp_per_capita) suggests that higher GDP per capita is associated with higher happiness scores.
#### A negative correlation between happiness_rank and happiness_score is expected, as lower ranks represent higher happiness.

## 04 Scatter plots

### 4.1 scatterplot for the happiness_score and economy_(gdp_per_capita)

```python
# Create a scatterplot for the happiness_score and economy_(gdp_per_capita)
sns.lmplot(x = 'happiness_score', y = 'economy_(gdp_per_capita)', data = df)
plt.title("Scatter Plot between happiness_score and economy_(gdp_per_capita)")
plt.tight_layout()
plt.show()
```

####  happiness_score and economy_(gdp_per_capita) has a positive correlation of 0.78, which shows the economy plays a vital role in the happiness score of that country

## 06 Categorical Plot

```python
# Create GDP bins
df['gdp_group'] = pd.cut(df['economy_(gdp_per_capita)'], 
                         bins=[0, 0.5, 1.0, 1.5, 2.0, df['economy_(gdp_per_capita)'].max()],
                         labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

# Create a boxplot to analyze happiness scores by GDP group
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='gdp_group', y='happiness_score', palette='viridis')
plt.title("Happiness Scores by GDP Group")
plt.xlabel("GDP Group")
plt.ylabel("Happiness Score")
plt.show()
```

### Interpretation of the Output:
### happiness Scores by GDP Group:

#### Analyze the variation in happiness scores across different GDP categories.
#### Higher GDP groups are likely to show higher median happiness scores, but outliers may indicate exceptions.
#### This plot can help us assess the economic impact on happiness.
